```python
class Solution:
    def judgePoint24(self, cards: List[int]) -> bool:

        # å› ç‚ºæµ®é»æ•¸ä¸èƒ½ç²¾ç¢ºè¡¨ç¤ºæ‰€æœ‰å°æ•¸ï¼Œæ‰€ä»¥æ¯”è¼ƒæ•¸å­—æ™‚ä¸é©åˆç”¨ ==
        # è€Œæ˜¯ç”¨ã€Œå·®çš„çµ•å°å€¼å°æ–¼ä¸€å€‹å®¹å¿å€¼ EPSILONã€ã€‚
        EPSLION = 1e-6

        def solve(nums):
            # Base case: if only one number is left
            if len(nums) == 1:
                return abs(nums[0] - 24) < EPSLION

            # Pick any two numbers i and j
            for i in range(len(nums)):
                for j in range(len(nums)):
                    if i == j:
                        continue

                    # Create a new list with the temaining numbers
                    new_list = [nums[k] for k in range(len(nums)) if k != i and k != j]

                    # Pergorm operations
                    a, b = nums[i], nums[j]

                    # Addition
                    if solve(new_list + [a + b]): return True

                    # Subtraction
                    if solve(new_list + [a - b]): return True
                    if solve(new_list + [b - a]): return True

                    # Mutiplication
                    if solve(new_list + [a * b]): return True

                    # Division (a/b and b/a)
                    if b != 0 and solve(new_list + [a / b]): return True
                    if a != 0 and solve(new_list + [b / a]): return True

                # If no combination worked 
                return False

        # Start the recursion with float versions of the cards
        return solve([float(c) for c in cards])
```

```python
import math, itertools

class Solution:
    def judgePoint24(self, nums: list[int]) -> bool:
        # èª¤å·®ç¯„åœï¼šè™•ç†æµ®é»æ•¸ç²¾åº¦èª¤å·®
        EPSILON = 1e-6

        # çµ¦å®šå…©å€‹æ•¸å­— a, bï¼Œç”¢ç”Ÿæ‰€æœ‰å¯èƒ½çš„é‹ç®—çµæœ
        def generate(a: float, b: float) -> list[float]:
            results = [
                a + b,   # a + b
                a - b,   # a - b
                b - a,   # b - aï¼ˆæ¸›æ³•è¦è€ƒæ…®é †åºï¼‰
                a * b    # a * b
            ]
            if b != 0:  
                results.append(a / b)   # a Ã· bï¼ˆé¿å…é™¤ä»¥ 0ï¼‰
            if a != 0:  
                results.append(b / a)   # b Ã· aï¼ˆé¿å…é™¤ä»¥ 0ï¼‰
                
            return results

        # DFS éè¿´å‡½æ•¸ï¼šä¸æ–·å˜—è©¦çµ„åˆæ•¸å­—ç›´åˆ°åªå‰©ä¸€å€‹
        def dfs(nums: list[float]) -> bool:
            # çµ‚æ­¢æ¢ä»¶ï¼šåªå‰©ä¸‹ä¸€å€‹æ•¸å­—
            if len(nums) == 1:
                # åˆ¤æ–·æ˜¯å¦æ¥è¿‘ 24ï¼ˆç”¨ EPSILON æ§åˆ¶èª¤å·®ï¼‰
                return abs(nums[0] - 24.0) < EPSILON

            # å¾ç›®å‰æ•¸åˆ—ä¸­æŒ‘å…©å€‹æ•¸å­—å‡ºä¾†ï¼ˆæ‰€æœ‰çµ„åˆï¼‰
            for i, j in itertools.combinations(range(len(nums)), 2):
                # å°é€™å…©å€‹æ•¸å­—åšæ‰€æœ‰å¯èƒ½é‹ç®—
                for num in generate(nums[i], nums[j]):
                    # å»ºç«‹ä¸‹ä¸€è¼ªæ•¸åˆ—ï¼š
                    # ç”¨é‹ç®—çµæœå–ä»£åŸæœ¬çš„å…©å€‹æ•¸ï¼Œå…¶ä»–æ•¸å­—ä¿ç•™
                    nextRound = [num] + [nums[k] for k in range(len(nums)) if k != i and k != j]
                    # éè¿´ç¹¼çºŒè™•ç†ä¸‹ä¸€è¼ª
                    if dfs(nextRound):
                        return True
            # å¦‚æœæ‰€æœ‰å¯èƒ½æ€§éƒ½å˜—è©¦éäº†é‚„æ˜¯ä¸è¡Œ â†’ å›å‚³ False
            return False

        # å¾åŸå§‹è¼¸å…¥é–‹å§‹ DFS
        return dfs(nums)
```

```python
class Solution:
    def judgePoint24(self, cards: list[int]) -> bool:
        def dfs(nums):
            if len(nums) == 1:
                return abs(nums[0] - 24) < 1e-6
            n = len(nums)
            for i in range(n):
                for j in range(i+1, n):
                    num1, num2 = nums[i], nums[j]
                    nextNums = [nums[k] for k in range(n) if k != i and k != j]
                    for val in (num1+num2, num1-num2, num2-num1, num1*num2):
                        if dfs(nextNums + [val]):
                            return True
                    if abs(num2) > 1e-6 and dfs(nextNums + [num1/num2]):
                        return True
                    if abs(num1) > 1e-6 and dfs(nextNums + [num2/num1]):
                        return True
            return False
        return dfs(list(map(float, cards)))
```
